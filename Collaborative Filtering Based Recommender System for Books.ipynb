# Rattings of the book along with book name
**Here we combine the book ratings with name as in our rating data set we just have ISBN**

**After merging we have Book-Title and other columns too**

```python
rattings_with_name = ratings.merge(books, on = "ISBN")
```

```python
rattings_with_name
```

### here we want to find the number of ratings per book that's why we do groupby using book-tile(not using ISBN bcz same name of a book exists with different ISBN ) 
the first column is book title and the number is Book ratings on that book (bcs we getting the this bcs we acquired this column by writting Book-Rating)

```python
rattings_with_name.groupby("Book-Title").count()['Book-Rating']
```

### lets call the above df as the num_rating_df
reset_index will convert the above data into the data frame 

```python
num_rating_df = rattings_with_name.groupby("Book-Title").count()['Book-Rating'].reset_index()
num_rating_df
```

### In the second col, we have number of rattings on that book so it's better to rename it

```python
num_rating_df.rename(columns = {"Book-Rating":"num_Ratings"}, inplace = True)
num_rating_df
```

```python
avg_rating_df = rattings_with_name.groupby("Book-Title").mean()['Book-Rating'].reset_index()
avg_rating_df
```

### Now again change the column name to the average Rating

```python
avg_rating_df.rename(columns = {"Book-Rating":"Avg-Rating"}, inplace = True)
avg_rating_df
```

### Here we merge two df number of ratings on a book along with avg rating

```python
popular_df = num_rating_df.merge(avg_rating_df, on = "Book-Title")
popular_df
```

## as we will be considered only those books who have 200 ratings on them 

```python
popular_df=popular_df[popular_df["num_Ratings"]>=200].sort_values('Avg-Rating', ascending = False).head(50)
```

## In the popular df, we just considered or having only 50 books that are more popular in dataset
**We use head fn thats why we get 50 books in the df**

```python
popular_df.head()
```

## To display it in the website we need more details too so we are going to merge it

```python
popular_df = popular_df.merge(books, on = "Book-Title").drop_duplicates('Book-Title')[['Book-Title','Book-Author','Publisher','Image-URL-L']]
popular_df
```

# Colaborative filtering
here we are getting only those user who rated more than 200 books

```python
X = rattings_with_name.groupby("User-ID").count()["Book-Rating"]>200
```

#### Here we get the index of those user how give more than 200 ratings 

```python
Readers_index = X[X].index
```

#### Here we get a bolean series lets put it into the df

```python
rattings_with_name["User-ID"].isin(Readers_index)
```

```python
filtered_ratings = rattings_with_name[rattings_with_name["User-ID"].isin(Readers_index)]
filtered_ratings.head()
```

```python
filtered_ratings.shape
```

```python
y = filtered_ratings.groupby("Book-Title").count()['Book-Rating'] >=50
```

```python
famous_book = y[y].index
```

```python
famous_book
```

```python
final_rattings = filtered_ratings[filtered_ratings['Book-Title'].isin(famous_book)]
```

```python
final_rattings.head()
```

## Converting it into pivot table in which at y side we will have books and on x side we have user with ratings given to the books by User

```python
pivot_table = final_rattings.pivot_table(index = 'Book-Title', columns = "User-ID", values = "Book-Rating")
pivot_table.head()
```

```python
pivot_table.fillna(0, inplace = True)
```

```python
pivot_table.shape
```

```python
pivot_table.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pivot_table)
similarity_score[0]
```

```python
similarity_score.shape
```

```python
def recommend(book_name):
    index = np.where(pivot_table.index == book_name)[0][0]
    similar_item = sorted(list(enumerate(similarity_score[index])), key = lambda x:x[1], reverse = True)[1:6]
    for i in similar_item:
        print(pivot_table.index[i[0]])
    
```

```python
recommend("1984")
```
